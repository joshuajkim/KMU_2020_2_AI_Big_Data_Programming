```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

### Question 1: Sum of CANCELLED of each WEEKDAY and each AIRLINE

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])[['CANCELLED']].sum()
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])[['CANCELLED']].agg('sum')
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY']).agg({'CANCELLED':'sum'})
```

### Question 2: % of CANCELLED of each WEEKDAY and each AIRLINE

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])[['CANCELLED']].mean()
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])[['CANCELLED']].agg('mean')
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY']).agg({'CANCELLED':'mean'})
```

### Question 3: Sum/% of DIVERTED, Sum/% of CANCELLED of each WEEKDAY and each AIRLINE

```python
flights.groupby(['AIRLINE','WEEKDAY'])[['DIVERTED', 'CANCELLED']].agg(['sum', 'mean'])
```

```python
flights.groupby(['AIRLINE','WEEKDAY']).agg({'DIVERTED':['sum','mean'], 'CANCELLED':['sum','mean']})
```

### Question 4: Count of ROWS, Sum/% of CANCELLED, Avg/Var of AIR_TIME of each ORG_AIR and each DEST_AIR

```python
Group_col = ['ORG_AIR', 'DEST_AIR']
Agg_dict = {'CANCELLED':['size', 'sum', 'mean'],
            'AIR_TIME':['mean', 'var']}
air_info = flights.groupby(Group_col).agg(Agg_dict)
air_info
```

## dealing with MultiIndex: columns -> rows

```python
air_info.columns
```

```python
level0 = air_info.columns.get_level_values(0)
level0
```

```python
level1 = air_info.columns.get_level_values(1)
level1
```

```python
level = level0 + '_' + level1
level
```

```python
air_info.columns = level
air_info
```

```python
air_info.reset_index(drop=False)
air_info
```
