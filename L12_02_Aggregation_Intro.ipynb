```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

# How was it delayed on average for each airline company?

### method 1

```python
# Series
flights.groupby('AIRLINE')['ARR_DELAY'].mean().round(2)
```

```python
# DataFrame
flights.groupby('AIRLINE')[['ARR_DELAY']].mean().round(2)
```

### method 2

```python
flights.groupby('AIRLINE')[['ARR_DELAY']].agg(np.mean).round(2)
```

### method 3 (IMPORTANT!!!!!!!)

```python
flights.groupby('AIRLINE').agg({'ARR_DELAY':'mean'})
```
