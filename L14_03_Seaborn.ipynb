```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white') # After sns's set_style, and then followed by below! Be careful!
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
%matplotlib inline
```

```python
sns.color_palette() # Current palette
```

```python
# Change the current palette to SUMMER!
sns.set_palette('summer')
# sns.set_palette(sns.color_palette('summer')): same
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

### Pair Plot

```python
sns.pairplot(data=iris)
;
```

```python
sns.pairplot(data=iris,
            hue='species')
# group by 'species'
;
```

```python
sns.pairplot(data=iris,
            vars=['sepal_length', 'petal_width'],
            hue='species')
;
```

```python
# regression line!
sns.pairplot(data=iris,
            vars=['sepal_length', 'petal_width'],
            kind='reg')
;
```

### Count Plot

```python
sns.countplot(data=titanic, x='survived')
;
```

```python
sns.countplot(data=titanic, x='sex',
             hue='survived')
;
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 5))

sns.countplot(data=titanic, x='sex', hue='survived', ax=ax[0])
sns.countplot(data=titanic, x='sex', ax=ax[1])
;
```

### Bar Plot

```python
sns.barplot(data=titanic, x='sex', y='survived',
           hue='class')
;
```
