```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

### User-defined Functions

* DEP_DELAY & ARR_DELAY: -100 ~ +60 (normal)
* what is the % of 'normal' of each AIRLINE?

```python
def is_normal(eachSeries):
    return eachSeries.between(-100, 60).mean()
```

```python
flights.groupby('AIRLINE')[['DEP_DELAY', 'ARR_DELAY']].agg(is_normal)
```

```python
flights.groupby('AIRLINE')[['DEP_DELAY', 'ARR_DELAY']].agg(['mean', is_normal])
```
