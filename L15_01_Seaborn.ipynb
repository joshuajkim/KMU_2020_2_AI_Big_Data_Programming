```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# boxplot: x-axis should be Categorical column!
sns.boxplot(data=titanic, x='survived', y='age')
;
```

```python
sns.boxplot(data=titanic, x='survived', y='age', hue='sex')
;
```

```python
sns.violinplot(data=titanic, x='survived', y='age')
;
```

```python
sns.violinplot(data=titanic, x='survived', y='age', hue='sex')
;
```

```python
# if you wanna see male and female are merged into one violine, then try this!
sns.violinplot(data=titanic, x='survived', y='age', hue='sex', split=True)
;
```

```python
sns.stripplot(data=titanic, x='survived', y='age', hue='sex')
;
```

```python
sns.stripplot(data=titanic, x='survived', y='age', hue='sex', dodge=True)
;
```

```python
# when you enter 'jitter=True', then many datapoints which have same values are scattered a little,
# so that you can recognize that at some range, there are many datapoints!
sns.stripplot(data=titanic, x='survived', y='age', hue='sex', dodge=True, jitter=True)
;
```

```python
sns.swarmplot(data=titanic, x='survived', y='age', hue='sex')
;
```

```python
sns.swarmplot(data=titanic, x='survived', y='age', hue='sex', dodge=True)
;
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width')
;
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='kde')
;
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='hex')
;
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', hue='species', kind='kde')
;
```

```python
sns.distplot(iris.sepal_length)
;
```

```python
sns.distplot(iris.sepal_length, hist=False)
;
```

```python
sns.distplot(iris.sepal_length, hist=False, rug=True)
;
```

```python
sns.distplot(iris.sepal_length, hist=True, kde=False, rug=True, color='red')
;
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 8), sharex=True)

for i in range(3):
    sns.distplot(titanic.loc[titanic['pclass'] == i + 1, 'fare'], ax=ax[i])
;
```

```python
sns.kdeplot(iris.sepal_width)
sns.kdeplot(iris.sepal_length)
;
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length)
;
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, shade=True)
;
```

```python
sns.lmplot(data=iris, x='sepal_width', y='sepal_length', hue='species', fit_reg=True)
;
```
