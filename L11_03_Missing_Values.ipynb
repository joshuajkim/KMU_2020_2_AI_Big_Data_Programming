```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('data/movie.csv')
movie.head()
```

```python
movie.info()
```

```python
# how many null values?
movie['duration'].isnull().sum()
```

```python
# how much percentage of null values?
movie['duration'].isnull().mean()
```

```python
# how many null values in each column?
movie.isnull().sum()
```

```python
# how much percentage of null values in each column?
movie.isnull().mean()
```

```python
# how many null values throughout all the columns?
movie.isnull().sum().sum()
```

```python
# when a row has at least one null value, the entire row will be deleted!
movie.dropna()
```

```python
movie.dropna().shape
```

```python
# default: how='any'
# but when you wanna delete a row where every datapoint is null value, then try how='all'!
movie.dropna(how='all')
```

```python
movie.dropna(how='all').shape
```

```python
# reset the index
movie.dropna().reset_index(drop=True)
# without 'drop=True', the original index will be located at the first column, so remember to write drop=True!
```

```python
# fill the null values with MODE!
movie['color'].value_counts()
```

```python
movie['color'].fillna('color')
```

```python
# or fill the null values with 'Unknown'
movie['color'].fillna('Unknown')
```

```python
# fill with MEAN!
movie['duration'].fillna(movie['duration'].mean())
```

```python
# fill with MEDIAN!
movie['duration'].fillna(movie['duration'].median())
```

```python
# Automatically fill all the null values with their medians only for numeric columns!
movie.fillna(movie.median())
```
