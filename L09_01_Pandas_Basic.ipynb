```python
# When you read the csv file, you can directly set the indices from one of columns!
movie2_df = pd.read_csv('data/movie.csv', index_col='movie_title')
movie2_df.head()
```

```python
# resetting the indices
    # when you reset the indices, the original index column moves to the 1st column. (Be careful!)
    # BUT, if you don't need the original index column anymore, you can write 'drop=True'!
movie3_df = movie2_df.reset_index()
movie3_df.head()
```

```python
# WOW!!!!! It's magic!
movie_df.dropna().reset_index(drop=True)
```

```python
# Renaming the indices and column titles!
movie_df.rename(index={0:'Avatar'}, columns={'color':'COLOUR', 'director_name':'DIRECTOR_NANE', 'duration':'LENGTH'})
```

```python
# renaming with this independent dict will be more convenient!
columns_dict = {'color':'COL', 'director_name':'DIRECTOR', 'duration':'DUR'}
movie_renamed = movie_df.rename(columns=columns_dict)
movie_renamed.head()
```

```python
# renaming with this list will be also more convenient as well!
columns_list = movie_df.columns.to_list()
columns_list[0] = 'COLOR'
columns_list[1] = 'DIRECTOR'
columns_list[3] = 'DURATION'
movie_df.columns = columns_list
movie_df.head()
```
