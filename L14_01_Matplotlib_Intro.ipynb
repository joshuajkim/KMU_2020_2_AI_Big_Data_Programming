```python
import matplotlib.pyplot as plt
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
# When an error happens because of the Korean language, try following!
# utf-8 -> euc-kr -> cp949 -> mbcs
weather = pd.read_csv('data/weather.csv', encoding='euc-kr')
weather.head()
```

```python
weather.columns = ['City', 'Month', 'Avg. Temp', 'Min Temp', 'Max Temp']
```

```python
seoul = weather[weather['City'] == '서울']
daegu = weather[weather['City'] == '대구']
```

```python
# figure to fig, axes to ax
fig, ax = plt.subplots()
# month to x axis, avg temp to y axis
ax.plot(seoul['Month'], seoul['Avg. Temp'])
# When you enter a semicolon, some dirty languages are deleted!
;
```

```python
# When you wanna put multiple series together in the same axes, then try this!
fig, ax = plt.subplots()
ax.plot(seoul['Month'], seoul['Avg. Temp'])
ax.plot(daegu['Month'], daegu['Avg. Temp'])
;
```

```python
# Style like below
fig, ax = plt.subplots()
ax.plot(seoul['Month'], seoul['Avg. Temp'],
       marker='o', linestyle='--', color='green')
;
```

```python
# axis labels and title
fig, ax = plt.subplots()

ax.plot(seoul['Month'], seoul['Avg. Temp'],
       marker='o', linestyle='--', color='green')

ax.set_xlabel('Month')
ax.set_ylabel('Avg. Temp')
ax.set_title('Monthly Temp in Seoul, Republic of Korea')
;
```

```python
# Multiple axes!
fig, ax = plt.subplots(2, 1, sharey=True) # sharey means: share y-axis scale together! wow!

ax[0].plot(seoul['Month'], seoul['Max Temp'],
          marker='o', linestyle='--', color='red')
ax[0].plot(seoul['Month'], seoul['Min Temp'],
          marker='o', linestyle='--', color='blue')
ax[0].plot(seoul['Month'], seoul['Avg. Temp'],
          marker='None', linestyle='-', color='black')

ax[1].plot(daegu['Month'], daegu['Max Temp'],
          marker='o', linestyle='--', color='red')
ax[1].plot(daegu['Month'], daegu['Min Temp'],
          marker='o', linestyle='--', color='blue')
ax[1].plot(daegu['Month'], daegu['Avg. Temp'],
          marker='None', linestyle='-', color='black')

ax[0].set_xlabel('Month')

ax[0].set_ylabel('Temp')
ax[1].set_ylabel('Temp')

ax[0].set_title('Temperature: Seoul vs Daegu')
;
```

### Bar Chart

```python
ol = pd.read_csv('data/olympics.csv', index_col='Country')
ol.head()
```

```python
fig, ax = plt.subplots()

ax.bar(ol.index, ol['Gold'],
      color='gold')

ax.set_title('Olympic Games')

ax.set_xlabel('Country')
ax.set_xticklabels(ol.index, rotation=90)

ax.set_ylabel('Number of Gold Medals')
;
```

```python
fig, ax = plt.subplots()

ax.bar(ol.index, ol['Gold'],
      color='gold', label='Gold',)
ax.bar(ol.index, ol['Silver'],
      color='gray', label='Silver',
      bottom=ol['Gold'])
ax.bar(ol.index, ol['Bronze'],
      color='brown', label='Bronze',
      bottom=ol['Gold']+ol['Silver'])

ax.set_title('olympic games')

ax.set_xlabel('Country')
ax.set_xticklabels(ol.index, rotation=90)

ax.set_ylabel('Number of Medals')

ax.legend()
;
```
