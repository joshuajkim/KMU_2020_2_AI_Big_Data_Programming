```python
# Filtering
# Operators in Pandas DataFrame
    # AND: &
    # OR: |
    # NOT: ~
    # XOR: ^
    
# i.e., Color is B/W and Duration is less than or equal to 60
f1 = movie2_df['color'] == 'Black and White'
f2 = movie2_df['duration'] <= 60

movie2_df[f1 & f2].head()
```

```python
# filtering with loc
movie2_df.loc[f1 & f2, ['director_name', 'country']]
```

```python
# filtering with iloc
movie2_df.iloc[(f1 & f2).values, :5] # iLoc can understand Series, so try ".values"! be careful!
```

```python
# Filtering method 1
%timeit movie2_df[movie2_df['color'] == 'Black and White']
# Filtering method 2: much faster than method 1
%timeit movie2_df.set_index('color').sort_index().loc['Black and White']
```

```python
# OR filtering method 1
f1 = movie2_df['language'] == 'Korean'
f2 = movie2_df['language'] == 'Chinese'
f = f1 | f2 # OR
movie2_df[f].head()
```

```python
# OR filtering method 2: ISIN
f_list = ['Korean', 'Chinese']
f = movie2_df['language'].isin(f_list)
movie2_df[f].head()
```

```python
# filtering method: 'BETWEEN'
f = movie2_df['duration'].between(60, 90) # inclusive!
movie2_df[f].head()
```

```python
# ISIN + BETWEEN
    # i.e., Korean + Chinese + Duration between 60 and 90
language_list = ['Korean', 'Chinese']

f1 = movie2_df['language'].isin(language_list)
f2 = movie2_df['duration'].between(60, 90)
f = f1 & f2
movie2_df[f].head()
```

```python
# SQL style code is also available in Python! WOW!
language_list = ['Korean', 'German']
query = 'language in @language_list and 60<=duration<=90' # @ means it is followed by a variable name!
movie2_df.query(query)
```

# Delete Duplicates!

```python
sample_dict = {'k1':['one'] * 3 + ['two'] * 4,
               'k2': [1,1,2,3,3,4,4]}
table_df = pd.DataFrame(sample_dict)
table_df
```

```python
table_df.drop_duplicates()
```

```python
# drop duplicates in a specific column!
table_df.drop_duplicates(['k1']) # it keeps the firstly found items!
```

```python
table_df.drop_duplicates(['k1'], keep='last') # it keeps the lastly found items!
```
